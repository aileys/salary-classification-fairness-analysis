```python
import numpy as np
import pandas as pd
from IPython.display import display

df1 = pd.read_csv("salary_test_data_male.csv")
df2 = pd.read_csv("salary_test_data_female.csv")

merged = pd.concat([df1,df2])

for race in merged["race"].unique():
  new_df = merged[merged["race"]==race]
  new_df.to_csv(f"salay_data_{race}.csv")
```

```python
test_df_aie = pd.read_csv("/content/salay_data_Amer-Indian-Eskimo.csv")
test_df_api = pd.read_csv("/content/salay_data_Asian-Pac-Islander.csv")
test_df_blk = pd.read_csv("/content/salay_data_Black.csv")
test_df_other = pd.read_csv("/content/salay_data_Other.csv")
test_df_white = pd.read_csv("/content/salay_data_White.csv")
```

```python
feature_columns = ["age","education-num","marital-status",
                                "occupation","hours-per-week",
                                "native-country","y"]

aie_data = test_df_aie[feature_columns]
api_data = test_df_api[feature_columns]
blk_data = test_df_blk[feature_columns]
other_data = test_df_other[feature_columns]
white_data = test_df_white[feature_columns]

display(aie_data.head())
display(api_data.head())
display(blk_data.head())
display(other_data.head())
display(white_data.head())
```

```python
from re import X
def featurize(df):
    X = df[['age','education-num','hours-per-week']]
    print (X)
    X['marital-status'] = [1 if x=='Married-civ-spouse' else 0 for x in df['marital-status']]
    X['native-country'] = [1 if x=='United-States' else 0 for x in df['native-country']]

    occ_groups = {
        'Priv-house-serv':0,'Other-service':0,'Handlers-cleaners':0,
        'Farming-fishing':1,'Machine-op-inspct':1,'Adm-clerical':1,
        'Transport-moving':2,'Craft-repair':2,'Sales':2,
        'Armed-Forces':3,'Tech-support':3,'Protective-serv':3,
        'Prof-specialty':4,'Exec-managerial':4,'?':0}

    X['occupation'] = [occ_groups.get(x.strip()) for x in df['occupation']]
    return X

# we also convery Y into 0/1 format

X_test_aie = featurize(aie_data)
y_actual_aie = [1 if y == '>50K' else 0 for y in aie_data['y']]
X_test_api = featurize(api_data)
y_actual_api = [1 if y == '>50K' else 0 for y in api_data['y']]
X_test_blk = featurize(blk_data)
y_actual_blk = [1 if y == '>50K' else 0 for y in blk_data['y']]
X_test_other = featurize(other_data)
y_actual_other = [1 if y == '>50K' else 0 for y in other_data['y']]
X_test_white = featurize(white_data)
y_actual_white = [1 if y == '>50K' else 0 for y in white_data['y']]

display(X_test_aie.head())
display(X_test_api.head())
display(X_test_blk.head())
display(X_test_other.head())
display(X_test_white.head())
```

```python
import pickle
from sklearn.metrics import accuracy_score

model_file = open("salary_classifier.model","rb")
model = pickle.load(model_file)

y_predicted_aie = model.predict(X_test_aie)
y_predicted_api = model.predict(X_test_api)
y_predicted_blk = model.predict(X_test_blk)
y_predicted_other = model.predict(X_test_other)
y_predicted_white = model.predict(X_test_white)

print(y_predicted_aie[:5])
print(y_predicted_api[:5])
print(y_predicted_blk[:5])
print(y_predicted_other[:5])
print(y_predicted_white[:5])
```

```python
from sklearn.metrics import classification_report

print ("Printing classification report for Amer-Indian-Eskimo:")
print(classification_report(y_actual_aie,y_predicted_aie))

print ("Printing classification report for Asian-Pac-Islander:")
print(classification_report(y_actual_api,y_predicted_api))

print ("Printing classification report for Black:")
print(classification_report(y_actual_blk,y_predicted_blk))

print ("Printing classification report for Other:")
print(classification_report(y_actual_other,y_predicted_other))

print ("Printing classification report for White:")
print(classification_report(y_actual_white,y_predicted_white))
```

The "Other" group has the lowest f1 score and has fewer samples which would be potential bias. The "Other" and "Amer-Indian-Eskimo" have significantly lower f1 scores indicating potential underrepresentation. Lower precision and recall means that the models may incorrectly predict positive cases.
